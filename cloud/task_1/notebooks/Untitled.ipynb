```python
import aiohttp
import asyncio
import nest_asyncio
nest_asyncio.apply()
```

```python
async def worker(session, value):
    try:
        async with session.get(f"https://swapi.dev/api/people/{value}/") as r:
            r.raise_for_status()
            output = await r.json()
            client.index(index='sw', doc_type='people', id=value, body=output)
            print("succes")
    except Exception as e:
        print('Connection Error', str(e))





async def main(values):
    async with aiohttp.ClientSession() as session:
        output = await asyncio.gather(*[worker(session, value) for value in range(1, values)])
        return output
    
loop = asyncio.get_event_loop()
data = loop.run_until_complete(main(100))
```
